```python
# Objective

# The objective of this notebook is to learn how to visualise biological data using Matplotlib and Seaborn. Through practical examples involving gene expression, cell viability, patient characteristics, and molecular data, this notebook demonstrates how graphical representations can be used to explore patterns, compare experimental groups, identify relationships, and communicate scientific findings effectively.
```

```python
# Why is Data Visualization Important in Bioinformatics?

# Biological datasets often contain thousands of genes, proteins, or clinical variables that are difficult to interpret using raw numerical values alone. Data visualisation helps researchers identify trends, detect outliers, compare experimental groups, and explore relationships between variables.
# Matplotlib and Seaborn are among the most widely used Python libraries for scientific visualisation. They play an essential role in bioinformatics, genomics, transcriptomics, proteomics, and biomedical research by transforming complex datasets into meaningful graphical representations.
```

```python
# Importing the Matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
# Line Plots
```

```python
days = [0, 24, 48, 72]
fold_change = [1.0, 2.5, 4.0, 5.2]
plt.plot(days, fold_change, marker = "o")
plt.title("miR-299 expression after treatment")
plt.xlabel("Time (hour)")
plt.ylabel("Fold Change")
plt.grid()
plt.show()
```

```python
concentration = [0, 5, 10, 15, 20]
viability = [100, 85, 70, 55, 40]
plt.plot(concentration, viability, marker = "o")
plt.title("Cell viability after treatment")
plt.xlabel("Concentration")
plt.ylabel("Viability")
plt.grid()
plt.show()
```

```python
hours = [0, 24, 48, 72]
expression = [1, 2.1, 3.8, 5.0]
plt.plot(hours, expression, marker = "o")
plt.title("miR-299 expression over time")
plt.xlabel("Time (hours)")
plt.ylabel("Expression")
plt.grid()
plt.show()
```

```python
patients = [1,2,3,4,5]
gene_exp = [12,18,25,30,20]
plt.plot(patients, gene_exp, marker = "s")
plt.title("Gene expression in patients")
plt.xlabel("Patients")
plt.ylabel("Gene expression")
plt.grid()
plt.show()
```

```python
# Bar Graphs
```

```python
treatment = ["Control", "Etoposide", "Cisplatin"]
expression = [1.0, 3.5, 2.8]
plt.bar(treatment, expression)
plt.title("miR-299 expression")
plt.xlabel("Treatment")
plt.ylabel("Expression")
plt.show()
```

```python
dose = ["0", "5", "10", "20"]
viability = [100, 85, 65, 40]
plt.bar(dose, viability, width = 0.5)
plt.title("Cell viability")
plt.xlabel("Drug Concentration")
plt.ylabel("Viability (%)")
plt.grid(axis = "y")
plt.show()
```

```python
genes = ["ESR1", "HDAC4", "CDKN1A", "SP1"]
scores = [95, 82, 76, 68]
plt.barh(genes, scores)
plt.title("Gene expression")
plt.xlabel("Expression")
plt.ylabel("Genes")
plt.show()
```

```python
patients = ["P1", "P2", "P3", "P4", "P5"]
expression = [12, 18, 25, 30, 20]
plt.bar(patients, expression, width = 0.5)
plt.title("Gene expression across different patients")
plt.xlabel("Patients")
plt.ylabel("Gene expression")
plt.grid(axis = "y")
plt.show()
```

```python
# Scatter Plots
```

```python
gene_a = [12, 15, 18, 22, 30]
gene_b = [14, 16, 20, 27, 35]
plt.scatter(gene_a, gene_b, marker = "o", s = 50)
plt.title("Gene expression of Gene A vs Gene B")
plt.xlabel("Gene A")
plt.ylabel("Gene B")
plt.grid()
plt.show()
```

```python
age = [25, 30, 35, 40, 45]
expression = [50, 45, 40, 35, 30]
plt.scatter(age, expression, marker = "s", s = 50)
plt.title("Gene expression vs Age")
plt.xlabel("Age")
plt.ylabel("Gene expression")
plt.grid()
plt.show()
```

```python
concentration = [0, 5, 10, 15, 20]
viability = [100, 85, 70, 55, 40]
plt.scatter(concentration, viability, marker = "o")
plt.title("Cell viability vs Concentration")
plt.xlabel("Drug Concentration")
plt.ylabel("Cell Viability")
plt.grid()
plt.show()
```

```python
# Box Plots
```

```python
control = [1.0, 1.1, 0.9, 1.2, 1.0]
treated = [2.5, 3.0, 2.8, 3.2, 2.7]
plt.boxplot([control, treated])
plt.xticks([1, 2], ["Control", "Treated"])
plt.title("miR-299 Expression")
plt.ylabel("Fold Change")
plt.show()
```

```python
normal = [12, 15, 18, 20, 22]
tumour = [35, 40, 45, 50, 55]
plt.boxplot([normal, tumour])
plt.xticks([1, 2], ["Normal", "Tumour"])
plt.title("miR-299 Expression")
plt.ylabel("Fold change")
plt.show()
```

```python
gene_a = [15, 18, 22, 25, 28]
gene_b = [40, 42, 45, 50, 55]
gene_c = [10, 15, 20, 22, 25]

plt.boxplot([gene_a, gene_b, gene_c])
plt.title("Expression of different genes")
plt.xticks([1, 2, 3], ["Gene A", "Gene B", "Gene C"])
plt.ylabel("Gene Expression")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
control = [1.0, 1.1, 0.9, 1.2, 1.0]
low_dose = [1.5, 1.8, 1.6, 1.7, 1.9]
high_dose = [2.8, 3.0, 3.2, 2.9, 3.1]

plt.boxplot([control, low_dose, high_dose])
plt.title("Expression of miR-299 across different samples")
plt.xticks([1, 2, 3], ["Control", "Low Dose", "High Dose"])
plt.ylabel("Gene expression")
plt.show()
```

```python
# Histograms
```

```python
ages = [22,24,25,26,27,28,30,31,35,36,38,40,42,45,47]
plt.hist(ages, bins = 5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
```

```python
expression = [10,12,14,15,18,20,22,24,26,28,30,35]
plt.hist(expression, bins = 6)
plt.title("Expression Distribution")
plt.xlabel("Expression")
plt.ylabel("Frequency")
plt.show()
```

```python
cell_viability = [98,95,92,90,88,85,82,80,75,70,68,65,60]
plt.hist(cell_viability, bins = 10)
plt.title("Cell Viability Distribution")
plt.xlabel("Viability")
plt.ylabel("Frequency")
plt.show()
```

```python
# Importing Seaborn library
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Scatter Plot
```

```python
gene_a = [10,15,20,25,30]
gene_b = [12,18,22,30,36]
sns.set_theme()
sns.scatterplot(x=gene_a, y=gene_b)
plt.title("Gene Expression of Gene A & Gene B")
plt.xlabel("Gene A")
plt.ylabel("Gene B")
plt.show()
```

```python
# Line Plot
```

```python
hours = [0,24,48,72]
expression = [1,2,3.5,5]
sns.set_theme()
sns.lineplot(x = expression, y = hours, marker = "o")
plt.title("Expression of Gene over time")
plt.xlabel("Expression")
plt.ylabel("Time (hours)")
plt.show()
```

```python
# Bar Graph
```

```python
groups = ["Control","Treatment"]
fold_change = [1.0,3.2]
sns.set_theme()
sns.barplot(x = groups, y = fold_change, width = 0.5)
plt.title("Gene expression in Control vs Treated Groups")
plt.xlabel("Groups")
plt.ylabel("Fold change")
plt.grid(axis = "y")
plt.show()
```

```python
# Box Plot
```

```python
control = [1.0,1.2,1.1,0.9,1.0]
treated = [2.8,3.0,2.9,3.1,2.7]
sns.set_theme()
sns.boxplot(data = [control, treated])
plt.title("Expression of Gene across different samples")
plt.xticks([0,1], ["Control", "Treated"])
plt.ylabel("Fold change")
plt.show()
```

```python
import pandas as pd
data = pd.DataFrame({
    "ESR1": [12, 18, 30],
    "HDAC4": [25, 30, 45],
    "SP1": [18, 22, 40]
},
index=["P1", "P2", "P3"])
```

```python
# Heatmaps
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
sns.heatmap(data, annot = True, cmap = "coolwarm", linewidths = 1)
plt.title("Gene expression in patients")
plt.show()
```

```python
df = pd.DataFrame({
    "GeneA":[10,20,30,40,50],
    "GeneB":[15,25,35,45,55],
    "GeneC":[50,40,30,20,10]
})
print(df)
```

```python
# Correlation Heatmaps
```

```python
sns.heatmap(df.corr(), annot = True, cmap = "coolwarm", linewidths =1, vmin = -1, vmax = 1)
plt.title("Correlation Matrix")
plt.show()
```

```python
df = pd.DataFrame({
    "Age":[25,30,35,40,45],
    "BMI":[22,24,26,28,30],
    "GeneExp":[15,20,25,30,35]
})
print(df)
print(df.corr())
```

```python
sns.heatmap(df.corr(), annot = True, linewidth = 1, cmap = "viridis", vmin = -1, vmax = 1)
plt.title("Correlation Matrix")
plt.show()
```
