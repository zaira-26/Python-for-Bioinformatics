```python
sequence = ""
with open("brca.fasta") as file:
  for line in file:
    line = line.strip()
    if line == "":
      continue
    if line.startswith(">"):
      print("Header:", line)
    else:
      sequence = sequence + line
print("Combined sequence:", sequence)
print("Length of sequence:", len(sequence))

gc_count = 0
at_count = 0
for i in sequence:
  if i == "G" or i =="C":
    gc_count += 1
  else:
    at_count += 1
print("AT count:", at_count)
print("GC count:", gc_count)

gc_percentage = gc_count/len(sequence) * 100
print("GC percentage is:", round(gc_percentage, 3), "%")
at_percentage = at_count/len(sequence) * 100
print("AT percentage is:", round(at_percentage, 3), "%")

if gc_percentage > 50:
  print("GC rich sequence")
else:
  print("AT rich sequence")

reverse_sequence = sequence[::-1]
print("Reverse sequence is:", reverse_sequence)
```

```python
sequences = {}
current_header = ""
current_sequence = ""

with open("brca.fasta") as file:
  for line in file:
    line = line.strip()
    if line == "":
      continue
    if line.startswith(">"):
      if current_header != "":
        sequences[current_header] = current_sequence
      current_header = line
      current_sequence = ""
    else:
      current_sequence = current_sequence + line

if current_header != "":
  sequences[current_header] = current_sequence

for header, sequence in sequences.items():
  print(header)
  print("Sequence:", sequence)
  print()

length = len(sequence)

counts = {"A":0, "G":0, "C":0, "T":0}
for nucleotide in sequence:
  counts[nucleotide] += 1

print("Nucleotide counts:", counts)

GC_count = counts["G"] + counts["C"]
gc_percentage = GC_count / length * 100
print("GC percentage is:", round(gc_percentage, 3))

AT_count = counts["A"] + counts["T"]
at_percentage = AT_count / length * 100
print("AT percentage:", round(at_percentage, 4))

reverse_sequence = sequence[::-1]
print("Reverse sequence:", reverse_sequence)
```
