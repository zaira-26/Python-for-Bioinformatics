```python
def gc_count(sequence):
  count = 0
  for i in sequence:
    if i == 'G' or i == 'C':
      count = count + 1
  return count

gc_count("AGTCGTCACCTA")
```

```python
def gc_percentage(sequence):
  count = 0
  for i in sequence:
    if i == 'G' or i == 'C':
      count = count + 1
      length = len(sequence)
      percentage = (count / length) * 100
  return percentage

gc_percentage("ATCGTCAAGTCATCA")
```

```python
def count_t(sequence):
  count = 0
  for i in sequence:
    if i == 'T':
      count = count + 1
  return count

count_t("ATCGTTCA")
```

```python
def at_sequence(sequence):          # Calculating AT in the sequence
  count = 0
  for i in sequence:
    if i == 'A' or i == 'T':
      count = count + 1
    total_length = len(sequence)
    percentage = (count/total_length) * 100
  return percentage

at_sequence("ATCGGTAACGTAC")
```

```python
def validate(sequence):           # Checking valid or invalide sequence
    if len(sequence) >= 12:
      return "Valid sequence"
    else:
      return "Invalid sequence"

validate("ATCGTA")
```

```python
def mutation(sequence):          # Checking mutation risk
  count = 0
  for i in sequence:
    if i == 'A':
      count = count + 1
  if count > 4:
      return "High mutation risk"
  else:
      return "Low mutation risk"

mutation("ATGCAAGTACGTAAA")
```

```python
def gc_rich(sequence):             # Checking the sequence for GC rich or GC low
  count = 0
  for i in sequence:
    if i == 'G' or i == 'C':
      count = count + 1
  if count >= 6:
    return "GC rich sequence"
  else:
    return "Low GC content"

gc_rich("ACGACCGGTACGGTACGGTA")
```

```python
def sequence_report(sequence):
  length = len(sequence)
  print("The length of sequence is:", length)
  count_at = 0
  count_gc = 0
  for i in sequence:
    if i == 'A' or i == 'T':
      count_at = count_at + 1
    if i == 'G' or i == 'C':
      count_gc = count_gc + 1
  print("AT count", count_at)
  print("GC count", count_gc)

  if count_gc >= 10:
    return "GC rich sequence"

  else:
    return "Low GC sequence"

sequence_report("ACGTAACGTACGTCAA")
```

```python
def compare_sequences(seq1, seq2):
  print("Sequence 1 is:", seq1)
  print("Sequence 2 is:", seq2)
  length_1 = len(seq1)
  print("Length of sequence 1 is:", length_1)
  length_2 = len(seq2)
  print("Length of sequence 2 is:", length_2)
  if length_1 > length_2:
    print(length_1)
  else:
    print(length_2)
  if length_1 == length_2:
    return "Sequences are equal in length"
  else:
    return "Sequences are not equal in length"

compare_sequences("ATCGATCAG", "AGTCG")
```

```python
# Given DNA sequence:
# print sequence length
# print reversed sequence
# print GC count
# classify as GC rich or AT rich

dna_sequence = "ATGCGGTACGTCCGTTTAAACGTTACGATCA"
length = len(dna_sequence)
print("The length of DNA sequence is:", length)
reverse_sequence = dna_sequence[::-1]
print("The reverse sequence is:", reverse_sequence)
g_count = dna_sequence.count("G")
c_count = dna_sequence.count("C")
gc_count = g_count + c_count
print("GC count in the sequence is:", gc_count)

gc_percentage = (gc_count / length) * 100
print("GC percentage in the sequence is:", round(gc_percentage, 2), "%")

if gc_percentage > 50:
  print("This is a GC rich sequence.")
else:
  print("This is AT rich sequence.")
```

```python
# Given list of sequences:
# find longest sequence
# print sequence
# print length

dna_sequences = ["ATG", "GTCGAT", "TTGCAGTACGTA", "AGCTACGGTACCGTTACCGTAC", "AGGGCTAGTA"]
longest_sequence = ""
for i in dna_sequences:
  if len(i) > len(longest_sequence):
    longest_sequence = i

print("The longest DNA sequence is:", longest_sequence)
print("The length of the sequence is", len(longest_sequence))

shortest_sequence = dna_sequences[0]
for i in dna_sequences:
  if len(i) < len(shortest_sequence):
    shortest_sequence = i
print("The shortest sequence is:", shortest_sequence)
print(len(shortest_sequence))
```

```python
#Count:
#A  #T  #G  #C
# using dictionary.
# Print most frequent nucleotide.
mutation = {"A": 0, "G": 0, "C": 0, "T": 0}
sequence = "ATGCTAACGTACGTCAA"
for i in sequence:
  mutation[i] += 1
print(mutation)

if mutation["A"] > mutation["G"] and mutation["A"] > mutation["C"] and mutation["A"] > mutation["T"]:
  print("A is the most frequent nucleotide.")
elif mutation["G"] > mutation["A"] and mutation["G"] > mutation["C"] and mutation["G"] > mutation["T"]:
  print("G is the most frequent nucleotide.")
elif mutation["C"] > mutation["A"] and mutation["C"] > mutation["G"] and mutation["C"] > mutation["T"]:
  print("C is the most frequent nucleotide.")
else:
  print("T is the most frequent nucleotide.")
```

```python
# GC percentage calculator function
def gc_percentage(sequence):
  count_gc = 0
  for i in sequence:
    if i == "G" or i == "C":
      count_gc += 1
  gc_percentage = (count_gc/len(sequence)) * 100
  return gc_percentage

def at_percentage(sequence):
  count_at = 0
  for i in sequence:
    if i =="A" or i == "T":
      count_at += 1
  at_percentage = (count_at/len(sequence)) * 100
  return at_percentage


sequence = "AGTCGATCGAAAAA"
print("The GC percentage in the sequence is:",round(gc_percentage(sequence),2),"%")
print("The AT percentage in the sequence is:",round(at_percentage(sequence),2),"%")
```
