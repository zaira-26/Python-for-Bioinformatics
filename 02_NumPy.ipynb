```python
# Why is NumPy important in Bioinformatics?
# Modern bioinformatics involves analysing large biological datasets such as gene expression profiles, DNA sequences, protein measurements, and genomic variants. NumPy provides an efficient way to store, manipulate, and perform mathematical operations on these datasets using multidimensional arrays.
# Many popular scientific Python libraries, including Pandas, SciPy, scikit-learn, and Biopython, are built on top of NumPy. Therefore, understanding NumPy is an essential step towards performing advanced bioinformatics analyses and developing computational biology pipelines.
```

```python
# Indexing and Slicing
```

```python
arr = np.array([10, 20, 30, 40, 50])
print(arr[0])
print(arr[-1])
print(arr[1:4])
```

```python
patients = np.array([101, 102, 103, 104, 105, 106])
print(patients[2:5])
```

```python
scores = np.array([
    [80, 75],
    [78, 93],
    [52, 76]
])
print(scores)
print(scores[1])
print(scores[2,1])
print(scores.shape)
```

```python
ages = np.array([18, 22, 35, 45, 60, 70])
print(ages[(ages > 20) & (ages < 50)])
```

```python
exp = np.array([5, 15, 25, 35, 45, 55])
print(exp[(exp < 10) | (exp > 50)])
```

```python
ages = np.array([18, 22, 35, 45, 60, 70])
print(ages[(ages < 20) | (ages > 50)])
```

```python
exp = np.array([10, 20, 30, 40, 50])
print(exp.sum())
print(exp.mean())
print(exp.min())
print(exp.max())
print(exp.std())
print(np.median(exp))
```

```python
fold_change = np.array([2.5, 3.1, 4.8, 5.2, 6.7])
print(fold_change.mean())
print(fold_change.std())
print(fold_change[fold_change > 4])
```
